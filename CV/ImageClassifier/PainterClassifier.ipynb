```
from fastai.vision.all import *
from fastai.vision.widgets import *
```

```
modelPath = Path.cwd()/"drive/My Drive/Colab Notebooks/artists/export.pkl"
model = load_learner(modelPath, cpu=True)
btnUpload = widgets.FileUpload()
out = widgets.Output()
lblPrediction = widgets.Label()
```

```
def make_prediction(x):
  lblPrediction.value = ""
  img = PILImage.create(btnUpload.data[-1])
  out.clear_output()
  with out: display(img.to_thumb(256,256))
  pred, predIdx, proba = model.predict(img)
  lblPrediction.value = f'This painting was created by {pred} with the probability of {proba[predIdx]*100:.2f}%'
```

```
btnUpload.observe(make_prediction, names=['data'])
```

```
display(VBox([widgets.Label("Upload the painting"), btnUpload, out, lblPrediction]))
```
